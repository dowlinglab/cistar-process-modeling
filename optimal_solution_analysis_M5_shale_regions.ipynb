# Optimal solution visualization: NGL feed composition sensitivity
## ROK model: M5; CO$_2$ tax rate: USD 45/tonne CO$_2$e; Region: EF-1 to EF-12

```python
model_code = 5
c_tax_rate = 4.5e-2 # $/kg-CO2e
regions = ['Bakken','EF-1','EF-2','EF-3','EF-4','EF-5','EF-6','EF-7','EF-8','EF-9','EF-10','EF-11','EF-12','EF-Basin']
purge_fraction = 0.01
```

```python
from src.unit_initialization import create_model_and_return_optimization_results
from src.plotting_functions import (plot_outlet_flowrate_by_region_horizontal_stacked_bars,
                                    plot_outlet_flow_rate_by_component_vs_region_lines,
                                    plot_LHV_by_region_stacked_vertical_bars,
                                    plot_emissions_by_region_vertical_stacked_bars
                                   )
import pandas as pd
import matplotlib.pyplot as plt
from idaes.core.util.tables import create_stream_table_dataframe, arcs_to_stream_dict, stream_states_dict
```

```python
model_data_dict = {}
```

## Read-in flowsheet data

```python
init_file_name = "./initialization_files/CISTAR_unit_initialization_{}_M{}.json.gz".format('Bakken', model_code)
for region in regions:
    converged_file_name = "./initialization_files/CISTAR_solve_constrained_{}_M{}_purge_{}.json.gz".format('Bakken', model_code, round(purge_fraction,3))
    costing_file_name = "./initialization_files/CISTAR_solve_with_costing_{}_C_tax_{}_M{}_purge_{}.json.gz".format('Bakken', 0.0, model_code, round(purge_fraction,3))
    optimal_file_name = "./initialization_files/CISTAR_optimal_solution_{}_C_tax_{}_M{}_purge_{}_sequential_solve.json.gz".format(region, c_tax_rate, model_code, round(purge_fraction,3))
    m, HI_df_initialized, HI_df_optimal = create_model_and_return_optimization_results(init_file_name, 
                                                                                       converged_file_name,
                                                                                       costing_file_name,
                                                                                       optimal_file_name,
                                                                                       model_code = model_code,
                                                                                       region = region,
                                                                                       c_tax_rate = c_tax_rate)
    
    stream_df = create_stream_table_dataframe(arcs_to_stream_dict(m, descend_into=True))
    with pd.ExcelWriter('solution_data.xlsx', mode='a',if_sheet_exists='replace') as writer:
        stream_df.to_excel(writer, sheet_name='streams_M{}_tax={}_{}'.format(model_code,c_tax_rate,region))
        HI_df_initialized.to_excel(writer, sheet_name='HI_M{}_tax={}_{}_initialized'.format(model_code,c_tax_rate,region))
        HI_df_optimal.to_excel(writer, sheet_name='HI_M{}_tax={}_{}_optimal'.format(model_code,c_tax_rate,region))
        
    model_data_dict[region] = m
```

## Extract data from flowsheets

```python
optimal_T_R102_dict = {}
optimal_T_H104_dict = {}
optimal_T_H105_dict = {}
optimal_P_F101_dict = {}
optimal_T_H106_dict = {}
optimal_P_H106_dict = {}
optimal_P_F102_dict = {}
optimal_MSP_dict = {}
optimal_Qs_dict = {}
optimal_Qw_dict = {}
optimal_product_LHV_dict = {}
optimal_TAC_dict = {}
optimal_h2_rebate_dict = {}
downstream_emissions_dict = {}
upstream_emissions_dict = {}
liquid_hc_flow_dict = {}
liquid_outlet_flow_dict = {}
liquid_hc_component_flow_dict = {}
liquid_comp_dict = {}
LHV_dict = {}
for region,m in model_data_dict.items():
    liquid_hc_comp_dict = {}
    LHV_by_comp_dict = {}
    flow_by_comp_dict = {}
    for c in m.fs.liquid:
        flow_by_comp_dict[c] = m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]()
    liquid_hc_component_flow_dict[region] = flow_by_comp_dict
    total_liquid_flow_rate = sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() for c in m.fs.liquid)
    for c in m.fs.liquid:
        liquid_hc_comp_dict[c] = 100 * m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]()/total_liquid_flow_rate
        LHV_by_comp_dict[c] = m.fs.LHV_per_component_per_stream['liq_outlet',c]()* total_liquid_flow_rate
    LHV_dict[region] = LHV_by_comp_dict
    liquid_comp_dict[region] = liquid_hc_comp_dict
    liquid_outlet_flow_dict[region] = total_liquid_flow_rate
    liquid_hc_flow_dict[region] = sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() for c in m.fs.liquid)
    optimal_T_R102_dict[region] = m.fs.H103.outlet.temperature[0]()
    optimal_T_H104_dict[region] = m.fs.H104.outlet.temperature[0]()
    optimal_T_H105_dict[region] = m.fs.H105.outlet.temperature[0]()
    optimal_P_F101_dict[region] = m.fs.F101.liq_outlet.pressure[0]()
    optimal_T_H106_dict[region] = m.fs.H106.outlet.temperature[0]()
    optimal_P_H106_dict[region] = m.fs.H106.outlet.pressure[0]()
    optimal_P_F102_dict[region] = m.fs.F102.liq_outlet.pressure[0]()
    optimal_MSP_dict[region] = m.fs.min_sell_price()
    optimal_Qs_dict[region] = m.fs.Qs()
    optimal_Qw_dict[region] = m.fs.Qw()
    downstream_emissions_dict[region] = m.fs.downstream_emissions()
    upstream_emissions_dict[region] = m.fs.upstream_emissions()
    optimal_product_LHV_dict[region] = m.fs.fuel_energy() + m.fs.gas_energy() + m.fs.h2_energy()
    optimal_TAC_dict[region] = m.fs.TAC()
    optimal_h2_rebate_dict[region] = m.fs.H2_purge_sell_price()
```

## Display and save optimal solution and DoFs to .csv file

```python
values_for_table = ['ROK_model','Region','C-tax-rate','MSP', 'Downstream-em','Product-LHV','H2-rebate','TAC','T_R102','T_H104','T_H105','P_F101','T_H106','P_H106','P_F102','Qs','Qw']
optimal_data = pd.DataFrame(columns=values_for_table)
for region in regions:
    data_dict = {}
    data_dict['ROK_model'] = 'M{}'.format(model_code)
    data_dict['Region'] = region
    data_dict['C-tax-rate'] = c_tax_rate*1000
    data_dict['Downstream-em'] = round(downstream_emissions_dict[region],3)
    data_dict['MSP'] = optimal_MSP_dict[region]/1000
    data_dict['T_R102'] = round(optimal_T_R102_dict[region],3)
    data_dict['T_H104'] = round(optimal_T_H104_dict[region],3)
    data_dict['T_H105'] = round(optimal_T_H105_dict[region],3)
    data_dict['P_F101'] = round(optimal_P_F101_dict[region],3)
    data_dict['T_H106'] = round(optimal_T_H106_dict[region],3)
    data_dict['P_H106'] = round(optimal_P_H106_dict[region],3)
    data_dict['P_F102'] = round(optimal_P_F102_dict[region],3)
    data_dict['Qs'] = round(optimal_Qs_dict[region]*1000/3600,3)
    data_dict['Qw'] = round(optimal_Qw_dict[region]*1000/3600,3)
    data_dict['Product-LHV'] = round(optimal_product_LHV_dict[region]*1000/3600,3)
    data_dict['TAC'] = round(optimal_TAC_dict[region],3)
    data_dict['H2-rebate'] = round(optimal_h2_rebate_dict[region],3)
    optimal_data = pd.concat([optimal_data,pd.DataFrame([data_dict])],ignore_index=True)

optimal_data.to_csv('optimal_data_wrt_region.csv',mode='w')
```
