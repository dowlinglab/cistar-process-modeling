# Optimal solution visualization: ROK model choice sensitivity
## Region: Bakken; CO$_2$ tax rates: USD 0 - 410/tonne CO$_2$e; ROK model: M2, M3, M4, M5

```python
model_codes = [2, 3, 4, 5]
c_tax_rate = 4.5e-2 # $/kg-CO2e
region = 'Bakken'
purge_fraction = 0.01
```

```python
from src.unit_initialization import create_model_and_return_optimization_results
from src.plotting_functions import (plot_outlet_flowrate_by_ROK_model_horizontal_stacked_bars,
                                    plot_outlet_flow_rate_by_component_vs_ROK_model_lines,
                                    plot_C4_C9_olefin_outlet_flowrate_by_ROK_model_vertical_bars,
                                    plot_emissions_by_oligo_model_vertical_stacked_bars
                                   )
                                
                                
import pandas as pd
import matplotlib.pyplot as plt
from idaes.core.util.tables import create_stream_table_dataframe, arcs_to_stream_dict, stream_states_dict
```

```python
model_data_dict = {}
```

## Read-in flowsheet data

```python
for model_code in model_codes:
    init_file_name = "./initialization_files/CISTAR_unit_initialization_{}_M{}.json.gz".format(region, model_code)
    converged_file_name = "./initialization_files/CISTAR_solve_constrained_{}_M{}_purge_{}.json.gz".format(region, model_code, round(purge_fraction,3))
    if model_code == 5:
        costing_file_name = "./initialization_files/CISTAR_solve_with_costing_{}_C_tax_{}_M{}_purge_{}.json.gz".format(region, '0.0', model_code, round(purge_fraction,3))
    else:
        costing_file_name = "./initialization_files/CISTAR_solve_with_costing_{}_C_tax_{}_M{}_purge_{}.json.gz".format(region, c_tax_rate, model_code, round(purge_fraction,3))
    if model_code == 5:
        optimal_file_name = "./initialization_files/CISTAR_optimal_solution_{}_C_tax_{}_M{}_purge_{}_sequential_solve.json.gz".format(region, c_tax_rate, model_code, round(purge_fraction,3))
    else:
        optimal_file_name = "./initialization_files/CISTAR_optimal_solution_{}_C_tax_{}_M{}_purge_{}.json.gz".format(region, c_tax_rate, model_code, round(purge_fraction,3))
    
    m, HI_df_initialized, HI_df_optimal = create_model_and_return_optimization_results(init_file_name, 
                                                                                       converged_file_name,
                                                                                       costing_file_name,
                                                                                       optimal_file_name,
                                                                                       model_code = model_code,
                                                                                       region = region,
                                                                                       c_tax_rate = c_tax_rate)
    
    stream_df = create_stream_table_dataframe(arcs_to_stream_dict(m, descend_into=True))
    with pd.ExcelWriter('solution_data.xlsx', mode='a',if_sheet_exists='replace') as writer:
        stream_df.to_excel(writer, sheet_name='streams_M{}_tax={}_{}'.format(model_code,c_tax_rate,region))
        HI_df_initialized.to_excel(writer, sheet_name='HI_M{}_tax={}_{}_initialized'.format(model_code,c_tax_rate,region))
        HI_df_optimal.to_excel(writer, sheet_name='HI_M{}_tax={}_{}_optimal'.format(model_code,c_tax_rate,region))
    model_data_dict['M{}'.format(model_code)] = m
```

## Extract data from flowsheets

```python
liquid_hc_list = ['butene','pentene','hexene','heptene','octene','nonene']
optimal_T_R102_dict = {}
optimal_T_H104_dict = {}
optimal_T_H105_dict = {}
optimal_P_F101_dict = {}
optimal_T_H106_dict = {}
optimal_P_H106_dict = {}
optimal_P_F102_dict = {}
optimal_MSP_dict = {}
optimal_Qs_dict = {}
optimal_Qw_dict = {}
optimal_product_LHV_dict = {}
optimal_TAC_dict = {}
optimal_h2_rebate_dict = {}
downstream_emissions_dict = {}
upstream_emissions_dict = {}
liquid_hc_flow_dict = {}
liquid_C4_C9_percent_dict = {}
liquid_hc_component_flow_dict = {}
for model_code,m in model_data_dict.items():
    flow_by_comp_dict = {}
    for c in m.fs.liquid:
        flow_by_comp_dict[c] = m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]()
    liquid_hc_component_flow_dict[model_code] = flow_by_comp_dict
    liquid_C4_C9_percent_dict[model_code] = 100*sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() for c in liquid_hc_list)/sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() for c in m.fs.liquid)
    liquid_hc_flow_dict[model_code] = sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() for c in m.fs.liquid)
    optimal_T_R102_dict[model_code] = m.fs.H103.outlet.temperature[0]()
    optimal_T_H104_dict[model_code] = m.fs.H104.outlet.temperature[0]()
    optimal_T_H105_dict[model_code] = m.fs.H105.outlet.temperature[0]()
    optimal_P_F101_dict[model_code] = m.fs.F101.liq_outlet.pressure[0]()
    optimal_T_H106_dict[model_code] = m.fs.H106.outlet.temperature[0]()
    optimal_P_H106_dict[model_code] = m.fs.H106.outlet.pressure[0]()
    optimal_P_F102_dict[model_code] = m.fs.F102.liq_outlet.pressure[0]()
    optimal_MSP_dict[model_code] = m.fs.min_sell_price()
    optimal_Qs_dict[model_code] = m.fs.Qs()
    optimal_Qw_dict[model_code] = m.fs.Qw()
    downstream_emissions_dict[model_code] = m.fs.downstream_emissions()
    upstream_emissions_dict[model_code] = m.fs.upstream_emissions()
    optimal_product_LHV_dict[model_code] = m.fs.fuel_energy() + m.fs.gas_energy() + m.fs.h2_energy()
    optimal_TAC_dict[model_code] = m.fs.TAC()
    optimal_h2_rebate_dict[model_code] = m.fs.H2_purge_sell_price()
    print(m.fs.TAC())
    print(sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() for c in m.fs.liquid))
    print(m.fs.MW_upstream_total['liq_outlet']())
    print('TAC/mass of liq HC = ',(m.fs.TAC())*1e8/(sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() 
                                                  for c in m.fs.liquid)*m.fs.MW_upstream_total['liq_outlet']()*3600*24*365)
         ,'USD/tonne')
    print('\n')
```

## Display and save optimal solution and DoFs to .csv file

```python
values_for_table = ['ROK_model','Region','C-tax-rate','MSP', 'Downstream-em','Product-LHV','H2-rebate','TAC','T_R102','T_H104','T_H105','P_F101','T_H106','P_H106','P_F102','Qs','Qw']
optimal_data = pd.DataFrame(columns=values_for_table)
for ROK_model in ROK_models_list:
    data_dict = {}
    data_dict['ROK_model'] = ROK_model
    data_dict['Region'] = region
    data_dict['C-tax-rate'] = c_tax_rate*1000
    data_dict['Downstream-em'] = round(downstream_emissions_dict[ROK_model],3)
    data_dict['MSP'] = optimal_MSP_dict[ROK_model]/1000
    data_dict['T_R102'] = round(optimal_T_R102_dict[ROK_model],3)
    data_dict['T_H104'] = round(optimal_T_H104_dict[ROK_model],3)
    data_dict['T_H105'] = round(optimal_T_H105_dict[ROK_model],3)
    data_dict['P_F101'] = round(optimal_P_F101_dict[ROK_model],3)
    data_dict['T_H106'] = round(optimal_T_H106_dict[ROK_model],3)
    data_dict['P_H106'] = round(optimal_P_H106_dict[ROK_model],3)
    data_dict['P_F102'] = round(optimal_P_F102_dict[ROK_model],3)
    data_dict['Qs'] = round(optimal_Qs_dict[ROK_model]*1000/3600,3)
    data_dict['Qw'] = round(optimal_Qw_dict[ROK_model]*1000/3600,3)
    data_dict['Product-LHV'] = round(optimal_product_LHV_dict[ROK_model]*1000/3600,3)
    data_dict['TAC'] = round(optimal_TAC_dict[ROK_model],3)
    data_dict['H2-rebate'] = round(optimal_h2_rebate_dict[ROK_model],3)
    optimal_data = pd.concat([optimal_data,pd.DataFrame([data_dict])],ignore_index=True)

optimal_data.to_csv('optimal_data_wrt_ROK_models.csv',mode='w')
```
