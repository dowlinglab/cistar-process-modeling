# Feed composition analysis
This notebook reads-in and visualizes shale feed compositions

```python
import pandas as pd
import numpy as np
```

```python
shale_inlet_df = pd.read_csv('Shale_composition.csv')
ng_frac_df = pd.read_csv('NGL_fraction.csv')
```

```python
shale_inlet_df
```

```python
import matplotlib.pyplot as plt
```

```python
regions = [c for c in shale_inlet_df if c != 'Species']
regions
```

```python
species_to_plot = ['methane','ethane','propane','nbutane','ibutane','pentane','hexane','heptane','octane']
# species_to_plot
```

```python
comp_dict = {}
for r in regions:
    region_comp_dict = {}
    for i,row in shale_inlet_df.iterrows():
        if row['Species'] in species_to_plot:
            region_comp_dict[row['Species']] = row[r]
    comp_dict[r] = region_comp_dict
```

```python
colors_dict = {'hydrogen':'lightgray',
                'methane':'gray',
                'ethane':'rosybrown',
                'propane':'yellowgreen',
                'nbutane':'tan',
                'ibutane':'orange',
                'pentane':'lime',
                'hexane':'paleturquoise',
                'heptane':'lightsteelblue',
                'octane':'violet',
                'ethylene': 'maroon',
                'propene': 'darkolivegreen',
                'butene': 'darkgoldenrod',
                'pentene': 'darkgreen',
                'hexene': 'lightseagreen',
                'heptene': 'royalblue',
                'octene': 'darkviolet',
                'nonene': 'crimson',
                }
```

```python
fig1, ax1 = plt.subplots(figsize=(8,6))
plt.rcParams.update({'font.size': 20})
left = np.zeros(len(regions))

for c,comp in comp_dict['EF-1'].items():
    p = ax1.barh(regions, [r[c]*100 for j,r in comp_dict.items()], label=c, left=left,color=colors_dict[c])
    
    left += [r[c]*100 for j,r in comp_dict.items()]
#     ax.bar_label(p, label_type='center',fontsize=8, fmt='%.2f')

ax1.set_xlabel('Feed mole percent [%]',fontsize=24,weight='bold')
ax1.set_ylabel('Region',fontsize=24,weight='bold')

ax1.tick_params(axis='y', labelsize= 16)
ax1.tick_params(axis='x', labelsize= 16)#, rotation=45)
# ax1.bar_label(p, label_type='edge',fontsize=16,weight='bold', fmt='%.1f')

ax1.set_xlim(0,100)
plt.legend(loc='upper center',bbox_to_anchor=(0.5, -0.2),ncol=3,fontsize=20)
plt.savefig('./plots/feed_compositions_bakken_eagleford.pdf',bbox_inches='tight',dpi=200)
plt.show()
```
