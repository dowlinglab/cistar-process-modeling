# Optimal solution visualization: CO$_2$ tax rate sensitivity
## ROK model: M5; Region: Bakken; CO$_2$ tax rates: USD 0 - 410/tonne CO$_2$e

```python
model_code = 5
c_tax_rates = [0.0,1e-5, 1e-3, 1.7e-2, 4.5e-2, 1.9e-1, 4.1e-1] # $/kg-CO2e
region = 'Bakken'
purge_fraction = 0.01
```

```python
from src.unit_initialization import create_model_and_return_optimization_results

from src.plotting_functions import (plot_MSP_vs_c_tax_rate
                                    )
import pandas as pd
import matplotlib.pyplot as plt
from idaes.core.util.tables import create_stream_table_dataframe, arcs_to_stream_dict, stream_states_dict
```

```python
model_data_dict = {}
```

## Read-in flowsheet data

```python
for c_tax_rate in c_tax_rates:
    init_file_name = "./initialization_files/CISTAR_unit_initialization_{}_M{}.json.gz".format(region, model_code)
    converged_file_name = "./initialization_files/CISTAR_solve_constrained_{}_M{}_purge_{}.json.gz".format(region, model_code, round(purge_fraction,3))
    costing_file_name = "./initialization_files/CISTAR_solve_with_costing_{}_C_tax_{}_M{}_purge_{}.json.gz".format(region, '0.0', model_code, round(purge_fraction,3))
    optimal_file_name = "./initialization_files/CISTAR_optimal_solution_{}_C_tax_{}_M{}_purge_{}_sequential_solve.json.gz".format(region, c_tax_rate, model_code, round(purge_fraction,3))
    m, HI_df_initialized, HI_df_optimal = create_model_and_return_optimization_results(init_file_name, 
                                                                                       converged_file_name,
                                                                                       costing_file_name,
                                                                                       optimal_file_name,
                                                                                       model_code = model_code,
                                                                                       region = region,
                                                                                       c_tax_rate = c_tax_rate)
    
    stream_df = create_stream_table_dataframe(arcs_to_stream_dict(m, descend_into=True))
    with pd.ExcelWriter('solution_data.xlsx', mode='a',if_sheet_exists='replace') as writer:
        stream_df.to_excel(writer, sheet_name='streams_M{}_tax={}_{}'.format(model_code,c_tax_rate,region))
        HI_df_initialized.to_excel(writer, sheet_name='HI_M{}_tax={}_{}_initialized'.format(model_code,c_tax_rate,region))
        HI_df_optimal.to_excel(writer, sheet_name='HI_M{}_tax={}_{}_optimal'.format(model_code,c_tax_rate,region))
        
    model_data_dict[c_tax_rate] = m
```

## Extract data from flowsheets

```python
liquid_hc_list = ['butene','pentene','hexene','heptene','octene','nonene']
optimal_T_R102_dict = {}
optimal_T_H104_dict = {}
optimal_T_H105_dict = {}
optimal_P_F101_dict = {}
optimal_T_H106_dict = {}
optimal_P_H106_dict = {}
optimal_P_F102_dict = {}
optimal_MSP_dict = {}
optimal_Qs_dict = {}
optimal_Qw_dict = {}
optimal_product_LHV_dict = {}
optimal_TAC_dict = {}
optimal_h2_rebate_dict = {}
downstream_emissions_dict = {}
upstream_emissions_dict = {}
liquid_hc_flow_dict = {}
liquid_C4_C9_percet_dict = {}
liquid_hc_component_flow_dict = {}
for c_tax_rate,m in model_data_dict.items():
    flow_by_comp_dict = {}
    for c in m.fs.liquid:
        flow_by_comp_dict[c] = m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]()
    liquid_hc_component_flow_dict[c_tax_rate] = flow_by_comp_dict
    liquid_C4_C9_percet_dict[c_tax_rate] = 100*sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() for c in liquid_hc_list)/sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() for c in m.fs.liquid)
    liquid_hc_flow_dict[c_tax_rate] = sum(m.fs.F102.liq_outlet.flow_mol_phase_comp[0,'Liq',c]() for c in m.fs.liquid)
    optimal_T_R102_dict[c_tax_rate] = m.fs.H103.outlet.temperature[0]()
    optimal_T_H104_dict[c_tax_rate] = m.fs.H104.outlet.temperature[0]()
    optimal_T_H105_dict[c_tax_rate] = m.fs.H105.outlet.temperature[0]()
    optimal_P_F101_dict[c_tax_rate] = m.fs.F101.liq_outlet.pressure[0]()
    optimal_T_H106_dict[c_tax_rate] = m.fs.H106.outlet.temperature[0]()
    optimal_P_H106_dict[c_tax_rate] = m.fs.H106.outlet.pressure[0]()
    optimal_P_F102_dict[c_tax_rate] = m.fs.F102.liq_outlet.pressure[0]()
    optimal_MSP_dict[c_tax_rate] = m.fs.min_sell_price()
    optimal_Qs_dict[c_tax_rate] = m.fs.Qs()
    optimal_Qw_dict[c_tax_rate] = m.fs.Qw()
    downstream_emissions_dict[c_tax_rate] = m.fs.downstream_emissions()
    upstream_emissions_dict[c_tax_rate] = m.fs.upstream_emissions()
    optimal_product_LHV_dict[c_tax_rate] = m.fs.fuel_energy() + m.fs.gas_energy() + m.fs.h2_energy()
    optimal_TAC_dict[c_tax_rate] = m.fs.TAC()
    optimal_h2_rebate_dict[c_tax_rate] = m.fs.H2_purge_sell_price()
```

## Plot MSP vs CO$_2$ tax rate

```python
xlabel_string = 'CO{} emissions tax rate \n [USD/tonne CO{}e]'.format('$_2$','$_2$')
ylabel_string = 'Mininum Selling Price \n [USD/MJ fuel]'
file_name_string = 'msp_vs_C-tax_rate_{}_M5.pdf'.format(region)
plot_MSP_vs_c_tax_rate(optimal_MSP_dict,c_tax_rates,xlabel_string,
                      ylabel_string,file_name_string)
```

## Display and save optimal solution and DoFs to .csv file

```python
values_for_table = ['ROK_model','Region','C-tax-rate','MSP', 'Downstream-em','Product-LHV','H2-rebate','TAC','T_R102','T_H104','T_H105','P_F101','T_H106','P_H106','P_F102','Qs','Qw']
optimal_data = pd.DataFrame(columns=values_for_table)
for c_tax_rate in c_tax_rates:
    data_dict = {}
    data_dict['ROK_model'] = 'M{}'.format(model_code)
    data_dict['Region'] = region
    data_dict['C-tax-rate'] = c_tax_rate * 1000
    data_dict['Downstream-em'] = round(downstream_emissions_dict[c_tax_rate],3)
    data_dict['MSP'] = optimal_MSP_dict[c_tax_rate]/1000
    data_dict['T_R102'] = round(optimal_T_R102_dict[c_tax_rate],3)
    data_dict['T_H104'] = round(optimal_T_H104_dict[c_tax_rate],3)
    data_dict['T_H105'] = round(optimal_T_H105_dict[c_tax_rate],3)
    data_dict['P_F101'] = round(optimal_P_F101_dict[c_tax_rate],3)
    data_dict['T_H106'] = round(optimal_T_H106_dict[c_tax_rate],3)
    data_dict['P_H106'] = round(optimal_P_H106_dict[c_tax_rate],3)
    data_dict['P_F102'] = round(optimal_P_F102_dict[c_tax_rate],3)
    data_dict['Qs'] = round(optimal_Qs_dict[c_tax_rate]*1000/3600,3)
    data_dict['Qw'] = round(optimal_Qw_dict[c_tax_rate]*1000/3600,3)
    data_dict['Product-LHV'] = round(optimal_product_LHV_dict[c_tax_rate]*1000/3600,3)
    data_dict['TAC'] = round(optimal_TAC_dict[c_tax_rate],3)
    data_dict['H2-rebate'] = round(optimal_h2_rebate_dict[c_tax_rate],3)
    optimal_data = pd.concat([optimal_data,pd.DataFrame([data_dict])],ignore_index=True)

optimal_data.to_csv('optimal_data_wrt_c_tax_rates.csv',mode='w')
```
